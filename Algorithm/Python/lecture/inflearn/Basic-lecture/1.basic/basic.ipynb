```python
# 1부터 N까지 홀수 출력
for i in range(1, 11):
    if i % 2 != 0:
        print(i, " ")

# 1부터 N까지의 합
sum = 0
for i in range(1, 11):
    sum += i
print(sum)

# N의 약수 출력하기 1 2 5 10
n = 10
for i in range(1, n + 1):
    if n % i == 0:
        print(i)
```

```python
# 중첩 반복문(2중 for문)
for i in range(5):
    for j in range(5):
        print(j, end=" ")
    print()

# 별찍기
for i in range(5):
    for j in range(i + 1):
        print("*", end=" ")
    print()

# 반대로 별찍기
for i in range(5):
    for j in range(5 - i):
        print("*", end=" ")
    print()
```

```python
# 문자열과 내장함수
msg = "It is Time"
print(msg.upper())
print(msg.lower())

tmp = msg.upper()
print(tmp)
print(tmp.find("T"))    # 인덱스 번호를 출력해준다.
print(tmp.count("T"))   # 원하는 문자열 캐릭터 카운트
print(msg)
print(msg[:2])          # 슬라이스 [start:end:increment]
print(msg[3:5])     
print(len(msg))         # 문자열 길이

for i in range(len(msg)):
    print(msg[i], end=" ")
print()

for i in msg:
    print(i, end=" ")
print()

for i in msg:
    if i.isupper():
        print(i, end=" ")
print()   
for i in msg:
    if i.islower():
        print(i, end=" ")
print()
for i in msg:
    if i.isalpha():
        print(i, end=" ")
print()

# 아스키 넘버 A(65) ~ Z(90) 대문자와 소문자의 넘버가 다르다.
temp = 'AZaz'
for x in temp:
    print(ord(x), end=" ")
print()

# 아스키 넘버 -> char
tmp = 65
print(chr(tmp))
```

```python
# 리스트와 내장함수(1)

import random as r

# range로 리스트 초기화
a = list(range(1, 11))
print(a)
b = [1, 2, 3, 4, 5]

# 리스트 합치기 (리스트를 직접 더하기, extend() 메서드 사용하기)
# c = a + b
# print(c)
# a.extend(b)
# print(a)

# value 추가
a.append(5)
print(a)
a.insert(3, 7) # 3번 인덱스에 7을 추가
print(a)

# value 삭제
## pop
c = a.pop()
print(a)
print(c)        # 꺼낸 값을 return
d = a.pop(3)    # 특정 인덱스 값을 pop가능
print(d)

## remove
a.remove(2)     # value를 파라미터로 받음, 인덱스가 아님
print(a)

# 인덱스 번호 출력
print(a.index(5))

a = list(range(1, 11))
print(a)
print(sum(a))
print(max(a))
print(min(a))       # 리스트 내의 값중에 최소값을 찾아준다.
print(min(7, 5))    # 인자값들 중에 최소값을 찾아준다.
print(a)
r.shuffle(a)        # 랜덤하게 섞어준다.
print(a)
a.sort()
print(a)
a.sort(reverse=True)
print(a)

# 리스트 내의 값을 모두 지운다.
a.clear()
print(a)

```

```python
# 리스트와 내장함수(2)

a = [23, 12, 36, 53, 19]
print(a[:3])
print(a[1:4])

b = []

# enumerate를 이용하면 리스트를 튜플로 값을 출력해준다. 인덱스와 함께 출력해준다.
# (0, 23) (1, 12) (2, 36) (3, 53) (4, 19)
for x in enumerate(a):
    print(x[0], x[1], end=" ")
print()

for index, value in enumerate(a):
    print(index, value, end= " ")
print()

# 리스트의 값 중 모두 참이라면 all 메서드는 true를 return 한다.
if all(60 > x for x in a):
    print("YES")
else:
    print("NO")

# 리스트의 값 중 하나라도 참이라면 any 메서드는 true를 return 한다.
if any(13 > x for x in a):
    print("YES")
else:
    print("NO")
```

```python
# 2차원 리스트 생성과 접근

# 같은 값으로 1차원 리스트 초기화
# a = [0] * 3

# 같은 값으로 2차원 리스트 초기화
a = [[0] * 3 for _ in range(3)]     # [0] * 3을 3번 돌린다는 의미
print(a)

for y in a:
    for x in y:
        print(x, end=' ')
    print()
```

```python
# 함수 만들기
def add(a, b):
    c = a + b
    print(c)
    return c, b     # 여러개의 값을 return 가능, 튜플로 리턴함

print(add(1, 2))

## 소수를 리턴하는 메서드
def isPrime(x):
    for i in range(2, x):
        # 2부터 자신의 이전의 값을 자신과 나눠서 나누어 떨어지는 값이 있다면 소수가 아니라고 판단함
        if x % i == 0:
            return False
    return True

a = [12, 13, 7, 9, 19]

for i in a:
    if isPrime(i):
        print(i, end=' ')
```

```python
# 람다 함수

def plus_one(x):
    return x + 1

print(plus_one(1))

plus_two = lambda x: x + 2  # 람다를 변수에 할당하여 사용한다.
print(plus_two(1))

a = [1, 2, 3]
print(list(map(plus_one, a)))           # map(func, list)
print(list(map(lambda x: x + 1, a)))    # 표현식을 func 인자에 람다식으로 바로 전달할 수 있다.

# 람다는 sort 할때 사용하면 유용하다.
```
