```python
# 풀이
"""
문제 풀이 핵심 아이디어
1. 스텍에 원소를 삽입할 때는 단순히 특정 수에 도다할 때까지 삽입하면 된다.
2. 스택에서 원소를 연달아 빼낼 때 내림차순을 유지할 수 있는지 확인한다.
"""

n = int(input())

count = 1
stack = []
result = []
b = True

for i in range(n):
    data = int(input())
    while count <= data:
        stack.append(count)     # count를 넣어서 오름차순으로 stack에 삽입한다.
        count += 1
        result.append("+")      
    if stack[-1] == data:
        stack.pop()
        result.append("-")
    else:
        print('NO')
        b = False
        break

if (b):
    print('\n'.join(result))
```

```python
# 백준 키로거 5397

# 풀이

test_case = int(input())

for _ in range(test_case):
    left_stack = []
    right_stack = []
    data = input()
    for i in data:
        if i == '-':
            if left_stack:
                left_stack.pop()
        elif i == '>':
            if right_stack:
                left_stack.append(right_stack.pop())
        elif i == '<':
            if left_stack:
                right_stack.append(left_stack.pop())
        else:
            left_stack.append(i)

    left_stack.extend(reversed(right_stack))
    print(''.join(left_stack))

        
```
